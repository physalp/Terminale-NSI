<h1><center><span style="color:#369;padding: 10px 30px 10px 30px;border: 3px solid #369;border-radius:7px;"> Listes chaînées </span></center></h1>




Une liste chaînée (ou liste liée) est une structure de données composées d’une séquence d’éléments de liste.
Chaque enregistrement d’une liste chaînée est appelé maillon.


##  <span style="color:#369"> I. Classe Maillon:</span>


Le champ de chaque maillon qui contient l’adresse du maillon suivant ou précédent est généralement appelé
pointeur.

Chaque élément (ou maillon) de la liste est composé :

• d’un contenu utile (de n’importe quel type),

• d’un pointeur pointant vers l’élément suivant de la séquence.

Le dernier élément de la liste possède un pointeur vide :





voici la class Maillon


```python
class Maillon:
    """Maillon d'une liste chainee"""
    def __init__(self,valeur=None,pointeur=None):
        self.valeur=valeur
        self.pointeur=pointeur

    def __str__(self):
        if self.pointeur==None:
            return f"{self.valeur} -> None"
        else:
            return f"{self.valeur} -> {str(self.pointeur)}"
    

```
