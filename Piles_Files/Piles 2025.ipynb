Qu'est-ce qu'une pile ?

```python
def pile():
    #retourne une liste vide
    return []


def vide(p):
    """Renvoie True si la pile est vide
    et False sinon"""
    return p==[]



def empiler(p,x):
    """Ajoute l’élément x à la pile p"""
    p.append(x)

    
def depiler(p):
    """Dépile et renvoie l’élément au sommet de la pile p"""
    assert not vide(p), "Pile vide"
    return p.pop()
```

##  <span style="color:#369"> I. Implémentation en objet:</span>



```python
class Pile:
    """
    Classe pile
    Création d'une instance Pile avec une liste
    """
    def __init__(self):
        """
        Initialisation d'une pile vide
        :return: La pile vide

        """
        self.L = []
    
    def __str__(self):
        """Méthode de représentation"""
        return 
        

    def vide(self) -> bool:
        """
        Teste si la pile est vide
        :return: bool: True si la pile est vide, False sinon
        """
        return self.L == []

    def depiler(self) -> object:
        """
        Dépile la pile
        :return: object: Le dernier élément de la pile
        """
        assert not self.vide(), "Pile vide"
        return self.L.pop()

    def empiler(self, x: object) -> None:
        """
        Empile un élément sur la pile
        :param x: object: L'élément à ajouter
        :return: None
        """
        self.L.append(x)
```

## __Exercice 2:__


Réaliser les méthodes taille(self) et sommet(self) qui retournent respectivement la
taille de la liste et le sommet de la liste ( sans le supprimer)

```python
class PileEx2(Pile):
    """ Classe qui hérite de la classe Pile : ceci permet de taper votre réponse dans cette cellule"""
    def __init__(self):
        Pile.__init__(self)
        

    def taille(self) -> int:
        """
        Renvoie la taille de la pile
        :return: int: La taille de la liste
        """
        return len(self.L)

    def sommet(self) -> object:
        """
        Renvoie le sommet de la pile (le dernier qui est rentré)
        :return: object: Le dernier élément rentré dans la pile
        """
        return self.L[self.taille()-1]
```

```python
p2 = PileEx2()
for i in range(5):
    p2.empiler(2 * i)


print(p2.taille())
print(p2.sommet())
```

on retourne faux )
